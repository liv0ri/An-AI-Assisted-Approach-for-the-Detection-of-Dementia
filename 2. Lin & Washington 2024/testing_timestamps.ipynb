- Checking the approcimate timestamps start and end for each word using the pickle folder. 
- Checking the actual start and end manually for each word.
- Checking whether the approximated start and end are the same as the expected start and end time.

## Insepect the word timestamps.

This is done by first checking 4 things:
1. Check 1: Is it a label we need?
2. Check 2: Have we already found a sample for this label?
3. Check 3: Is it the ORIGINAL transcript (not augmented)?
4. Check 4: Does it have enough words to inspect?

```python
def inspect_word_timestamps(pickle_path, num_words=5):
    with open(pickle_path, "rb") as f:
        data = pickle.load(f)

    labels_to_find = ["control", "dementia"]
    samples = {}

    for u in data:
        label = u.get('label')
        if (label in labels_to_find and 
            label not in samples and 
            u.get('source_type') == 'original' and 
            len(u.get('words', [])) >= num_words):
            
            samples[label] = u
        
        if len(samples) == len(labels_to_find):
            break

    if not samples:
        print("No valid original samples found. Check if 'source_type' is set to 'original' in your pickle.")
        return

    for label, sample_utt in samples.items():
        print(f"\n{'='*25} ORIGINAL {label.upper()} SAMPLE {'='*25}")
        print(f"Source File: {sample_utt['source_file']}")
        
        start_sec = sample_utt['start'] / 1000.0
        end_sec = sample_utt['end'] / 1000.0
        print(f"Utterance Range: {start_sec:.2f}s - {end_sec:.2f}s")

        for i in range(num_words):
            word = sample_utt['words'][i]
            start_t, end_t = sample_utt['word_times'][i]
            
            print(f"  Word {i+1}: '{word}'")
            print(f"    CHAT Time: {start_t/1000.0:.2f}s to {end_t/1000.0:.2f}s")

        if 'audio' in sample_utt:
            sr = 16000 
            actual_dur = len(sample_utt['audio']) / sr
            print(f"  Audio Buffer: {len(sample_utt['audio'])} samples (~{actual_dur:.2f}s)")

inspect_word_timestamps(PROCESSED_DATA_PATH)
```
