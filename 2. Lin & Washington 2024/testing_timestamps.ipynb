# Testing Timestamps

In this Juypter Notebook the relevance of the timestamps features will be evaluated. This will be done by:
- Taking five word in an audio file.
- Checking the approcimate timestamps start and end for each word using the pickle folder. 
- Checking the actual start and end manually for each word.
- Checking whether the approximated start and end are the same as the expected start and end time.


## Import the necessary libraries.

```python
import pickle
import numpy as np
from src.models_code.config import PROCESSED_DATA_PATH
import soundfile as sf
import os

PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
save_dir = os.path.join(PROJECT_ROOT, "Other", "a. Data Visualisation and Distribution")
```

## Insepect the word timestamps.

This is done by first checking 4 things:
1. Check 1: Is it a label we need?
2. Check 2: Have we already found a sample for this label?
3. Check 3: Is it the ORIGINAL transcript (not augmented)?
4. Check 4: Does it have enough words to inspect?

```python
def inspect_word_timestamps(pickle_path, num_words=5):
    with open(pickle_path, "rb") as f:
        data = pickle.load(f)

    labels_to_find = ["control", "dementia"]
    samples = {}

    for u in data:
        label = u.get('label')
        if (label in labels_to_find and 
            label not in samples and 
            u.get('source_type') == 'original' and 
            len(u.get('words', [])) >= num_words):
            
            samples[label] = u
        
        if len(samples) == len(labels_to_find):
            break

    if not samples:
        print("No valid original samples found. Check if 'source_type' is set to 'original' in your pickle.")
        return

    for label, sample_utt in samples.items():
        print(f"\n{'='*25} ORIGINAL {label.upper()} SAMPLE {'='*25}")
        print(f"Source File: {sample_utt['source_file']}")
        
        start_sec = sample_utt['start'] / 1000.0
        end_sec = sample_utt['end'] / 1000.0
        print(f"Utterance Range: {start_sec:.2f}s - {end_sec:.2f}s")

        for i in range(num_words):
            word = sample_utt['words'][i]
            start_t, end_t = sample_utt['word_times'][i]
            
            print(f"  Word {i+1}: '{word}'")
            print(f"    CHAT Time: {start_t/1000.0:.2f}s to {end_t/1000.0:.2f}s")

        if 'audio' in sample_utt:
            sr = 16000 
            actual_dur = len(sample_utt['audio']) / sr
            print(f"  Audio Buffer: {len(sample_utt['audio'])} samples (~{actual_dur:.2f}s)")

inspect_word_timestamps(PROCESSED_DATA_PATH)
```

## Visualise all datapoints

```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def visualize_dataset_stats(pickle_path):
    with open(pickle_path, "rb") as f:
        data = pickle.load(f)

    # Convert list of dicts to a DataFrame for easier analysis
    df = pd.DataFrame([{
        'label': u['label'],
        'source_type': u['source_type'],
        'num_words': len(u['words']),
        'duration_ms': u['end'] - u['start'],
        'source_file': u['source_file']
    } for u in data])

    plt.figure(figsize=(16, 10))
    sns.set_theme(style="whitegrid")

    # 1. Class Distribution (Original vs Augmented)
    plt.subplot(2, 2, 1)
    sns.countplot(data=df, x='label', hue='source_type')
    plt.title("Data Points per Class & Type")
    plt.ylabel("Count")

    # 2. Utterance Length (Word Count) Distribution
    plt.subplot(2, 2, 2)
    sns.histplot(data=df, x='num_words', hue='label', kde=True, element="step")
    plt.title("Distribution of Words per Utterance")
    plt.xlabel("Number of Words")

    # 3. Audio Duration Distribution
    plt.subplot(2, 2, 3)
    sns.kdeplot(data=df, x='duration_ms', hue='label', fill=True)
    plt.title("Utterance Duration (ms) by Class")
    plt.xlabel("Milliseconds")

    # 4. Data Volume by Source File (Top 20 files)
    plt.subplot(2, 2, 4)
    file_counts = df[df['source_type'] == 'original']['source_file'].value_counts().head(20)
    file_counts.plot(kind='bar')
    plt.title("Top 20 Files by Utterance Count (Original)")
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, "dataset_stats.png"))
    plt.show()

    # Print summary statistics
    print(f"--- Dataset Summary ---")
    print(df.groupby(['label', 'source_type']).size().unstack(fill_value=0))
    print(f"\nAverage words per utterance: {df['num_words'].mean():.2f}")
    print(f"Total Unique Source Files: {df['source_file'].nunique()}")


visualize_dataset_stats(PROCESSED_DATA_PATH)
```
